```python
import spacy;
import nltk;
import text_formatter;

from text_formatter import cleaned_text_getter;
from nltk.stem.porter import PorterStemmer;
from spacy.language import Language;
from spacy import displacy;

# Text preprocessing logic is abstracted within text_formatter.py
cleaned_text = cleaned_text_getter();
print(cleaned_text);


# Prevents 'gđo' from being recognized as the end of a sentence in spaCy.
@Language.component("gdjo_not_end_of_sent")
def gdjo_not_end_of_sent(doc):

    for t in doc:
        if (t.text == "gđo"):
            doc[t.i+2].is_sent_start = False;

    return doc;

# Initialize NLP processing for the Croatian language
nlp = spacy.load("hr_core_news_sm");

# Add the custom sentence boundary rule before the parser
nlp.add_pipe("gdjo_not_end_of_sent", before="parser");

# Process the cleaned text with spaCy
doc = nlp(cleaned_text);

# Split the text into sentences
for sentence in doc.sents:
    print(sentence);

# Extract complex sentences (longer than 20 words)
punctuation = [".", ",", "!", "?", ";"];
long_sentences = [];

for sentence in doc.sents:
    word_count = 0;
    for word in sentence:
        if (word.text not in punctuation):
            word_count = word_count + 1;
            if (word_count >= 20):
                long_sentences.append(sentence);
                break;

# Print long sentences
for sentence in long_sentences:
    print(sentence);

# Print word dependencies, parts of speech, and tags
for t in doc:
    print(t.text, t.dep_, t.pos_, t.tag_);

# Count the number of verbs and nouns
verb_count = 0;
noun_count = 0;

for t in doc:
    if (t.pos_ == "VERB"):
        verb_count = verb_count + 1;
    elif (t.pos_ == "NOUN"):
        noun_count = noun_count + 1;

# Compare the frequency of nouns and verbs  
if (verb_count > noun_count):
    print(f"Andrić uses more verbs ({verb_count}) than nouns ({noun_count}).");
elif (noun_count > verb_count):
    print(f"Andrić uses more nouns ({noun_count}) than verbs ({verb_count}).");
else:
    print(f"Andrić uses an equal number of nouns and verbs, totaling {noun_count} each.");

# Remove stop words from the text
non_stop_tokens = [token.text for token in doc if not token.is_stop]
text_with_no_stop_words = " ".join(non_stop_tokens);

print(text_with_no_stop_words);

# Perform stemming using the Porter Stemmer
p_stem = PorterStemmer(); 

for t in non_stop_tokens:
    print(t, "------>", p_stem.stem(t));

# Perform lemmatization
for token in doc:
    print(token.text, "---->", token.lemma_);

# Syntactic dependency visualization
displacy.render(doc, style="dep", jupyter=True, options={'distance' : 90});
displacy.render(doc.sents, style="dep", jupyter=True, options={"distance": 100});
displacy.serve(doc.sents, style="dep");

# Named entity recognition (NER)
for e in doc.ents:
    print(e.text, e.label_, spacy.explain(e.label_));
```
